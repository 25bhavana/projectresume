```python
def qualityupdate(df):
    for i,row in df.iterrows():
        val = row['quality']
        if val  <=6:
            df.at[i,'quality']=0
        else:
            df.at[i,'quality']=1
qualityupdate(df)
```

```python
df.head(3)
```

train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x= df.drop(['quality'],axis=1) # df.iloc[:,:-1]
y=df['quality'] # df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
```

```python
from sklearn.svm import SVC
```

```python
reg = SVC()
```

```python
reg.fit(xtrain,ytrain)
```

```python
reg.score(xtest,ytest)
```

```python
yp = reg.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
c = confusion_matrix(ytest,yp)
```

```python
c
```

```python
sns.heatmap(c)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(reg,{
    'C':[0.1,0.4,0.8,1.0,1.3],
    'gamma':[0.1,0.4,0.8,1.0,1.3],
    'kernel':['rbf','linear']
},scoring='accuracy',cv=10) 
model
```

```python
model.fit(xtrain,ytrain)
```

```python
model.best_params_
```

```python
mod = SVC(C=1.3,gamma=1.0,kernel='rbf')
```

```python
mod.fit(xtrain,ytrain)
```

```python
mod.score(xtest,ytest)
```

Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
rfc2 = cross_val_score(estimator=rfc,X=xtrain,y=ytrain,cv=10)
```

```python
rfc2.mean()
```
